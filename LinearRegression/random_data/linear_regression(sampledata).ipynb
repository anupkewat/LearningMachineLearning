```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0,9.0)
```

```python
#input data
data=pd.read_csv('data.csv')

df=pd.DataFrame(data=data)
```

```python
x=data.iloc[:,0]
y=data.iloc[:,1]
```

```python
#scatter plot
plt.scatter(x,y)
```

```python
#building the model
m=0
c=0
L= 0.0001
epochs= 10000 #iterations of gradient decsent
n=float(len(x))
costlist=[]
```

```python
#gradient decsent
for i in range(epochs):
    Y_pred=m*x+c
    cost=(1/(2*n))*np.sum(np.square(Y_pred-y))
    d_m=(-1/n)*sum(x*(y-Y_pred))#derovative wrt m
    d_c=(-1/n)*sum(y-Y_pred)#derivative wrt c
    m=m-L*d_m
    c=c-L*d_c
    costlist.append(cost)
print(m,c)
```

```python
Y_pred=m*x+c
plt.scatter(x,y)
plt.plot(x,Y_pred,color='orange')
plt.show()
```

```python
rng=np.arange(0,epochs)
plt.plot(costlist,rng)
plt.show
```
